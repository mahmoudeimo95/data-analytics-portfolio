```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/index_1.csv")
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])

df["card"] = df["card"].fillna("Cash Payment")

print("Data loaded successfully!")

print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(df.info())
```

```python
plt.figure(figsize=(8, 5))

df["coffee_name"].value_counts().plot(
    kind="bar",
    color="saddlebrown"
)

plt.title("Most Popular Coffee")
plt.xlabel("Coffee Type")
plt.ylabel("Number of Sales")
plt.xticks(rotation=30, ha="right")

plt.tight_layout()
plt.savefig("chart1_popular_coffee.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))

df.groupby("coffee_name")["money"].sum().sort_values().plot(
    kind="barh",
    color="peru"
)

plt.title("Revenue by Coffee Type")
plt.xlabel("Total Revenue")

plt.tight_layout()
plt.savefig("chart2_revenue_by_coffee.png")
plt.show()
```

```python
plt.figure(figsize=(6, 6))

df["cash_type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    colors=["#6F4E37", "#D2A679"]
)

plt.title("Payment Method Distribution")
plt.ylabel("")

plt.tight_layout()
plt.savefig("chart3_payment_methods.png")
plt.show()
```
